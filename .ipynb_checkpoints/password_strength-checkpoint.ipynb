```python
import pandas as pd
import numpy as np
```

```python
my_file = open("dataset.txt", encoding="utf8")
string_list = []
for i in my_file.readlines():
    string_list.append(i)
my_file.close()
```

```python
string_list[0] = string_list[0].replace(',',' ')
for i in range(1,len(string_list)):
    string_list[i] = string_list[i].replace(string_list[i][-3],' ')
```

```python
string_list
```

```python
file1 = open("dataset1.text","a")
for i in range(len(string_list)):
    file1.write(string_list[i])
file1.close()
```

```python
data = pd.read_csv('dataset.csv',',',error_bad_lines = False)
data.head()
```

```python
data.isnull().sum()
```

```python
#1 null value in password
data = data.dropna()
```

```python
data_array = np.array(data)
np.random.shuffle(data_array)
```

```python
x = [labels[0] for labels in data_array]
y = [labels[1] for labels in data_array]
y
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.countplot(x='strength',data = data)
```

```python
def string_to_chararray(inputs):
    chararray = []
    for i in inputs:
        chararray.append(i)
    return chararray
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer = string_to_chararray)
x = vectorizer.fit_transform(x)
```

```python
vectorizer.vocabulary_
```

```python
feature_names = vectorizer.get_feature_names()
first_document_vector = x[0]
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index = feature_names, columns=["tfidf"]) #todense for getting only tf-idf score
df.sort_values(by=["tfidf"],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
log_class = LogisticRegression(penalty='l2',multi_class='ovr')
log_class.fit(x_train,y_train)
```

```python
print(log_class.score(x_test,y_test))
```

```python
clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(x_train, y_train) #training
print(clf.score(x_test, y_test))
```

```python
x_predict = np.array(["ameya"])
x_predict = vectorizer.transform(x_predict)
y_pred = log_class.predict(x_predict)
print(y_pred)
```

```python
x_predict = np.array(["ameya1000"])
x_predict = vectorizer.transform(x_predict)
y_pred = log_class.predict(x_predict)
print(y_pred)
```

```python
x_predict = np.array(["ameya@#12"])
x_predict = vectorizer.transform(x_predict)
y_pred = log_class.predict(x_predict)
print(y_pred)
```
